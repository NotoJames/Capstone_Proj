```python
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pl.read_csv("player_rankings_pca.csv")

# drop player column
df = df.select(pl.all().exclude("player"))
df = df.select([
    "net_dpr",
    "nade_ratio",
    "headshotperhit",
    "engagement_rate",
    "survivability",
    "efficiency",
    "rounds_played",
    "clutches",
    "rounds_won",
    "match_wins",
    "matches_played",
    "clutch_rate",
    "win_rate",
    "match_win_rate"
])


# scale
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.to_numpy())

# linkage
linkage_data = linkage(df_scaled, method="complete", metric="euclidean")

results = []
cluster_size_data = {}

for k in range(2, 6):
    labels = fcluster(linkage_data, k, criterion="maxclust")
    
    # metrics
    sil = silhouette_score(df_scaled, labels)
    ch = calinski_harabasz_score(df_scaled, labels)
    db = davies_bouldin_score(df_scaled, labels)
    
    # cluster counts
    unique, counts = np.unique(labels, return_counts=True)
    cluster_counts = dict(zip(unique, counts))
    
    print(f"\n=== k = {k} ===")
    print("Cluster sizes:", cluster_counts)
    print(f"Silhouette: {sil:.4f}, CH: {ch:.2f}, DB: {db:.4f}")
    
    # store
    results.append({
        "n_clusters": k,
        "silhouette_score": sil,
        "calinski_harabasz": ch,
        "davies_bouldin": db
    })
    
    cluster_size_data[k] = cluster_counts

results_df = pd.DataFrame(results)
```
